```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure(figsize = (10, 10))
ax = Axes3D(fig)
X = np.arange(-10, 10)
Y = np.arange(-10, 10)
X, Y = np.meshgrid(X, Y)
Z = X * 5 + 100
Z1 = X * 5 + 120
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, Z1, colors='purple')
show()
```

```python
fig = figure(figsize = (10, 10))
ax = Axes3D(fig)
X = np.arange(-10, 10)
Y = []
for i in X:
    i = i * 5
    Y.append(i)
X, Y = np.meshgrid(X, Y)

Z = X ** 2 - Y ** 2 
Z1 = X ** 2 + Y ** 2

ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z)
show()
```
