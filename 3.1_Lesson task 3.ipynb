```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = []
x1 = []
y = []
y1 = []

for i in range(1000):
    r = 500
    x.append(-i)
    x1.append(i)
    y.append(np.sqrt(r ** 2 - i ** 2))
    y1.append(-np.sqrt(r ** 2 - i ** 2))

plt.figure(figsize = (6, 6)) 
plt.plot(x, y)
plt.plot(x, y1)
plt.plot(x1, y1)
plt.plot(x1, y)
plt.xlabel("x")
plt.ylabel("y")
```

```python
x = []
x1 = []
y = []
y1 = []

for i in range(1000):
    a = 80
    b = 120
    x.append(-i)
    x1.append(i)
    y.append(np.sqrt(b ** 2 - (i ** 2 / a ** 2) * (b **2)))
    y1.append(-np.sqrt(b ** 2 - (i ** 2 / a ** 2) * (b **2)))

plt.plot(x, y)
plt.plot(x, y1)
plt.plot(x1, y1)
plt.plot(x1, y)
plt.xlabel("x")
plt.ylabel("y")
```
