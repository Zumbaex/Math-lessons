```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
k = 3
k1 = 5
a = 1.5
a1 = 2.5
b = -4
b1 = 4
x = np.linspace (-2*np.pi, 3*np.pi, 201)
y = k*np.cos(x-a) + b
y1 = k1*np.cos(x-a1) + b1
y2 = k*np.cos(x-a1) + b
plt.plot(x, y)
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
Пусть точки A1(x1, y1) и A2(x2, y2) посредством ортогонального 
преобразования переводятся соответствено в точки A'1 (x'1, y'1) и A'2(x'2, y'2)
Требуется доказать, что отрезки A1A2 и A'1A'2 имеют равные длины

Доказательство:
|A'1A'2|^2 = [x'2-x'1]^2 + [y'2-y'1]^2 + [a11(x2-x1) + a12(y2-y1)]^2 +
+ [a21(x2-x1) + a22(y2-y1)]^2 = (a11^2 + a21^2)*(x2-x1)^2 + (a12^2 + a22^2)* 
*(y2-y1)^2 + 2(a11a12 + a21a22)*(x2-x1)*(y2-y1)=(x2-x1)^2 + (y2-y1)^2=|A1A2|^2

Т.о. |A1A2| = |A'1A'2|
```

```python
#exp(x) + x∙(1 – y) = 1  
#y = x2 – 1

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
plt.figure(figsize = (12, 6))
x = np.linspace(-2, 4, 201)
plt.plot(x, (np.exp(x) + x - 1) / x)
plt.plot(x, x**2 -1)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

def equations(p):
    x, y = p
    return(x**2 - 1 - y, np.exp(x) + x*(1-y) - 1)
x1, y1 = fsolve(equations, (-1.5, 4))
x2, y2 = fsolve(equations, (2.5, 8))
print(f'x1 = {round(x1, 2)}, y1 = {round(y1, 2)}')
print(f'x2 = {round(x2, 2)}, y2 = {round(y2, 2)}')
```

```python
#y = x2 – 1
#exp(x) + x∙(1 – y)  - 1 > 0
from scipy.optimize import fsolve
plt.figure(figsize = (8, 14))
x = np.linspace(-2, 4, 201)
plt.plot(x, x**2 - 1)
plt.plot(x, (-1/x) + 2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

def equations(p):
    x, y = p
    return(x**2 - 1 - y, (-1/x) + 2 - y)

x1, y1 = fsolve(equations, (-0.1, -2))
x2, y2 = fsolve(equations, (0.5, -5))
x3, y3 = fsolve(equations, (1.5, 4))
print(f'x1 = {round(x1, 2)}, y1 = {round(y1, 2)}')
print(f'x2 = {round(x2, 2)}, y2 = {round(y2, 2)}')
print(f'x3 = {round(x3, 2)}, y3 = {round(y3, 2)}')
```
